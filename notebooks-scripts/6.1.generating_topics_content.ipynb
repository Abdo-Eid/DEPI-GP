````python
import json
import jsonlines
import time
import random
import threading
from tqdm import tqdm


# -------------------------------
# GLOBAL RATE LIMIT
# -------------------------------
TOKENS_PER_MINUTE = 250_000
TOKENS_PER_REQUEST = 2012
REQUESTS_PER_MINUTE = TOKENS_PER_MINUTE // TOKENS_PER_REQUEST
REQUEST_INTERVAL = 60 / REQUESTS_PER_MINUTE

rate_lock = threading.Lock()
progress_lock = threading.Lock()
last_call_time = time.time()

# -------------------------------
# RATE LIMITER
# -------------------------------
def rate_limited_api_call():
    global last_call_time

    with rate_lock:
        now = time.time()
        elapsed = now - last_call_time

        if elapsed < REQUEST_INTERVAL:
            time.sleep(REQUEST_INTERVAL - elapsed)

        last_call_time = time.time()


# -------------------------------
# API CALL FUNCTION
# -------------------------------
def generate_text(model_name, sys_prompt, topic, subtopic,
                  temperature=0.85, max_tokens=8192):

    prompt = f"{topic}\n{subtopic}"

    for attempt in range(5):
        try:
            rate_limited_api_call()

            completion = client.chat.completions.create(
                model=model_name,
                messages=[
                    {"role": "system", "content": sys_prompt},
                    {"role": "user", "content": prompt}
                ],
                temperature=temperature,
                max_completion_tokens=max_tokens,
                top_p=0.95,
                reasoning_effort="low",
                stream=False
            )

            return completion.choices[0].message.content

        except Exception as e:
            if "429" in str(e) or "rate limit" in str(e).lower():
                sleep_time = 2 ** attempt + random.random()
                print(f"[Rate Limit] Retrying in {sleep_time:.2f}s")
                time.sleep(sleep_time)
            else:
                raise e

    raise RuntimeError("API failed after multiple retries")


# -------------------------------
# WORKER FUNCTION (WITH PROGRESS BAR)
# -------------------------------
def worker(worker_id, items, writer, sys_prompt, model_name, pbar):
    row_id = 1

    for topic, subtopics in items:

        # --- Topic == subtopic ---
        text = generate_text(model_name, sys_prompt, topic, topic)

        writer.write({
            "id": f"{worker_id}-{row_id}",
            "topic": topic,
            "subtopic": topic,
            "text": text
        })
        row_id += 1

        # Update progress bar safely
        with progress_lock:
            pbar.update(1)

        # --- Subtopics ---
        for sub in subtopics:
            text = generate_text(model_name, sys_prompt, topic, sub)

            writer.write({
                "id": f"{worker_id}-{row_id}",
                "topic": topic,
                "subtopic": sub,
                "text": text
            })
            row_id += 1

            # Update progress bar safely
            with progress_lock:
                pbar.update(1)

````

````python

# -------------------------------
# MAIN THREAD LAUNCHER
# -------------------------------
def process_with_two_threads(input_file, output_file, sys_prompt, model_name):

    # Load JSON data
    with open(input_file, "r", encoding="utf-8") as f:
        data = list(json.load(f).items())

    # Split for 2 workers
    midpoint = len(data) // 2
    part1 = data[:midpoint]
    part2 = data[midpoint:]

    # Total rows = topic + its subtopics
    total_rows = sum(1 + len(subs) for _, subs in data)

    writer = jsonlines.open(output_file, mode="w")

    # Shared progress bar
    pbar = tqdm(total=total_rows, desc="Processing", ncols=100)

    # Create threads
    t1 = threading.Thread(target=worker, args=(1, part1, writer, sys_prompt, model_name, pbar))
    t2 = threading.Thread(target=worker, args=(2, part2, writer, sys_prompt, model_name, pbar))

    t1.start()
    t2.start()

    t1.join()
    t2.join()

    pbar.close()
    writer.close()

    print("\nProcessing complete!")
````

````python
process_with_two_threads(
    input_file="/content/egy_agri_topics.json",
    output_file="/content/egy_agri_topics.jsonl",
    sys_prompt=SYS_PROMPT,
    model_name="gpt-oss-120b"
)
````

# ✅ **What the `id` Means (Briefly)**

The `id` has the format:

```
<worker_id>-<row_number>
```

Example:

```
1-34 → worker thread #1, row 34
2-12 → worker thread #2, row 12
```

This ensures:

* Unique IDs across both threads
* Easy debugging
* Avoiding collisions

---

# ✅ **Correct Order (Briefly)**

To sort the final JSONL correctly:

1. Split the `id` into two integers:

   * `worker_id`
   * `row_number`
2. Sort by:

```
(worker_id ASC, row_number ASC)
```

This recreates the original processing order.

````python
import json
import re

def find_missing_ids_using_original(original_json_file, jsonl_file):
    """
    Determines missing IDs by:
      1) Counting expected rows based on original JSON topics/subtopics
      2) Extracting all existing IDs from JSONL
      3) Returning the missing IDs

    Assumes ID format: "<worker_id>-<row_number>"
    """

    # -----------------------------
    # 1) Load original JSON structure
    # -----------------------------
    with open(original_json_file, "r", encoding="utf-8") as f:
        orig = json.load(f)

    expected_total = sum(1 + len(subs) for subs in orig.values())
    print(f"Expected total rows: {expected_total}")

    # -----------------------------
    # 2) Extract all IDs from JSONL
    # -----------------------------
    id_pattern = re.compile(r"(\d+)-(\d+)")
    existing_ids = set()

    with open(jsonl_file, "r", encoding="utf-8") as f:
        for line in f:
            try:
                row = json.loads(line)
                m = id_pattern.match(str(row.get("id", "")))

                if m:
                    worker = int(m.group(1))
                    num = int(m.group(2))
                    existing_ids.add((worker, num))

            except:
                continue  # skip corrupted lines

    print(f"Found {len(existing_ids)} existing IDs")

    # -----------------------------
    # 3) Determine expected IDs structure
    # -----------------------------
    # You used 2 workers → IDs start from worker 1 and 2
    # Split approximately evenly

    # How many rows per worker ideally?
    per_worker = expected_total // 2
    worker_ranges = {
        1: (1, per_worker),
        2: (1, expected_total - per_worker)
    }

    # Build expected ID set
    expected_ids = set()
    for worker, (start, end) in worker_ranges.items():
        for row_num in range(start, end + 1):
            expected_ids.add((worker, row_num))

    # -----------------------------
    # 4) Missing = expected − existing
    # -----------------------------
    missing_ids = sorted(expected_ids - existing_ids)

    return {
        "expected_total": expected_total,
        "found_total": len(existing_ids),
        "missing_count": len(missing_ids),
        "missing_ids": missing_ids
    }
````

````python
def map_missing_ids_to_pairs(
    original_json_file,
    missing_ids
):
    """
    Maps missing IDs (worker_id, row_number) to (topic, subtopic)
    using the same order the generator used.
    """

    # -----------------------------
    # Load original JSON topics
    # -----------------------------
    with open(original_json_file, "r", encoding="utf-8") as f:
        data = json.load(f)

    # -----------------------------
    # Build ordered list of (topic, subtopic)
    # -----------------------------
    ordered_pairs = []
    for topic, subs in data.items():
        ordered_pairs.append((topic, topic))  # topic == subtopic
        for s in subs:
            ordered_pairs.append((topic, s))

    total = len(ordered_pairs)

    # -----------------------------
    # Reconstruct worker assignment
    # -----------------------------
    assignments = {}  # mapping (worker_id, worker_row_index) → index in ordered_pairs

    worker_positions = {1: 1, 2: 1}  # row counters for each worker
    worker = 1  # alternate workers: 1, then 2, then 1 ...

    for i in range(total):
        topic, subtopic = ordered_pairs[i]

        worker_row = worker_positions[worker]
        assignments[(worker, worker_row)] = (topic, subtopic)

        worker_positions[worker] += 1
        worker = 2 if worker == 1 else 1  # switch worker

    # -----------------------------
    # Map missing IDs → topic/subtopic
    # -----------------------------
    missing_pairs = []
    for (worker_id, row_num) in missing_ids:
        if (worker_id, row_num) in assignments:
            missing_pairs.append({
                "worker": worker_id,
                "row": row_num,
                "topic": assignments[(worker_id, row_num)][0],
                "subtopic": assignments[(worker_id, row_num)][1]
            })
        else:
            missing_pairs.append({
                "worker": worker_id,
                "row": row_num,
                "topic": None,
                "subtopic": None,
                "error": "ID outside expected range"
            })

    return missing_pairs
````

````python
import jsonlines
from tqdm import tqdm

def regenerate_missing_pairs(
    missing_pairs,
    output_file,
    sys_prompt,
    model_name
):
    """
    One-threaded regeneration of ONLY the missing items,
    using their ORIGINAL ID (worker-row), with tqdm progress bar.
    """

    with jsonlines.open(output_file, mode="w") as writer:

        for item in tqdm(missing_pairs, desc="Regenerating missing rows"):
            worker = item.get("worker")
            rownum = item.get("row")
            topic = item.get("topic")
            subtopic = item.get("subtopic")

            # Skip unmapped cases
            if topic is None or subtopic is None:
                continue

            # Call your generate_text function
            text = generate_text(
                model_name=model_name,
                sys_prompt=sys_prompt,
                topic=topic,
                subtopic=subtopic
            )

            # Use ORIGINAL ID
            original_id = f"{worker}-{rownum}"

            writer.write({
                "id": original_id,
                "topic": topic,
                "subtopic": subtopic,
                "text": text
            })

    print(f"Finished regenerating {len(missing_pairs)} missing rows into {output_file}")
````

````python
regenerate_missing_pairs(
    missing_pairs=missing_pairs,
    output_file="egy_agri_topics_missing_fixed.jsonl",
    sys_prompt=SYS_PROMPT,
    model_name="gpt-oss-120b"
)
````

````python
def combine_and_sort_jsonl(original_jsonl, fixed_jsonl, output_jsonl):
    """
    Combines original + regenerated JSONL files,
    sorts them by (worker_id, row_number),
    and writes a clean final JSONL file.
    """

    id_pattern = re.compile(r"(\d+)-(\d+)")  # matches "1-44"

    all_rows = []

    # -----------------------------
    # Load original JSONL
    # -----------------------------
    with open(original_jsonl, "r", encoding="utf-8") as f:
        for line in f:
            try:
                row = json.loads(line)
                all_rows.append(row)
            except:
                continue  # skip corrupt lines

    # -----------------------------
    # Load fixed JSONL
    # -----------------------------
    with open(fixed_jsonl, "r", encoding="utf-8") as f:
        for line in f:
            try:
                row = json.loads(line)
                all_rows.append(row)
            except:
                continue

    # -----------------------------
    # Sort by worker + row_number
    # -----------------------------
    def id_sort_key(row):
        m = id_pattern.match(str(row["id"]))
        if not m:
            return (999999, 999999)  # push broken IDs to bottom
        worker, num = m.groups()
        return (int(worker), int(num))

    all_rows.sort(key=id_sort_key)

    # -----------------------------
    # Write final sorted file
    # -----------------------------
    with jsonlines.open(output_jsonl, mode="w") as writer:
        for row in all_rows:
            writer.write(row)

    print(f"Finished combining + sorting into {output_jsonl}")
````

````python
combine_and_sort_jsonl(
    original_jsonl="egy_agri_topics.jsonl",
    fixed_jsonl="egy_agri_topics_missing_fixed.jsonl",
    output_jsonl="egy_agri_topics_final_sorted.jsonl"
)
````
